Combining those files into one

```
read_files = glob.glob("/content/Language-Model/*.txt") # membaca semua file txt pada folder

with open("result.txt", "wb") as outfile: # menggabungkan semua file txt menjadi satu file hasil txt
    for f in read_files:
        with open(f, "rb") as infile:
            outfile.write(infile.read())
```

Checking the output (in lowercase to minimalize mistake)

```
data = open("result.txt","r").read()
textfile = data.lower() # mengubah semua huruf teks menjadi huruf kecil untuk meminimalisir kesalahan
textfile
```

text tokenization

```
sent_text = nltk.sent_tokenize(textfile)
print('token kalimat :',sent_text)

mod_sentences = [] # array/list untuk menampung semua kalimat teks awal

# loop per kalimat, tambahkan tag <s> di awal kalimat dan </s> di akhir kalimat
for sentence in sent_text:
    mod_sentence = []
    sent_tokens = nltk.word_tokenize(sentence)
    mod_sentence.append('<s>')
    for token in sent_tokens:
      mod_sentence.append(token)
    mod_sentence.append('</s>')
    mod_sentences.append(mod_sentence)

print('token kata/simbol per kalimat :',mod_sentences)
```

frequency counter (unigram)

```
freq_tab = {}
total_count = 0
for sentence in mod_sentences:
  for token in sentence:
    if token in freq_tab:
        freq_tab[token] += 1 # kata sudah ada di dictionary, update frekuensinya
    else:
        freq_tab[token] = 1 # kata belum ada di dictionary 
    total_count += 1
        
print('frekuensi model unigram :',freq_tab)
```

frequency leaderboard (top 10, unigram)

```
unigram_sort = dict(sorted(freq_tab.items(), key=operator.itemgetter(1),reverse=True)[:10]) # melakukan sorting 10 besar berdasarkan frekuensi
print('10 besar kata/simbol yang sering muncul (unigram) :',unigram_sort)
```

probability counter (unigram)

```
prob_tab = {}
for token in freq_tab:
    prob_tab[token] = freq_tab[token]/total_count
    
print('probabilitas setiap kata/simbol (unigram) :',prob_tab)
```

frequency counter (bigram)

```
freq_bigram_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])
    if curr_bigram in freq_bigram_tab:
      freq_bigram_tab[curr_bigram] += 1 # bigram sudah ada di dictionary, update frekuensinya
    else:
      freq_bigram_tab[curr_bigram] = 1 # bigram belum ada di dictionary

print('frekuensi model bigram :',freq_bigram_tab)
```

probability counter (bigram)

```
bigram_prob_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])

    if curr_bigram not in bigram_prob_tab:  
      bigram_prob_tab[curr_bigram] = freq_bigram_tab[curr_bigram]/freq_tab[sentence[i-1]] # bigram belum ada di dictionary, hitung probability  
 
print('probabilitas model bigram :',bigram_prob_tab)
```

probability leaderboard (top 10, bigram)

```
bigram_sort = dict(sorted(bigram_prob_tab.items(), key=operator.itemgetter(1),reverse=True)[:10]) # melakukan sorting 10 besar berdasarkan probabilitas
print('10 model bigram berdasarkan probabilitas tertinggi :',bigram_sort)
```
